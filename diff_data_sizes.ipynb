```python
_, deploy = evaluate(Egreedy(1))
rnd_ctr = deploy[-1]
```

```python
sizes = [100,30,20,10,5,1]
results = [['Size','Bucket','Algorithm','CTR Lift']]

algorithms = [Egreedy(0.1),ThompsonSampling(),Ucb1(0.1),LinUCB(0.3, context="both"), LinUCB(0.3, context="user")]

for alg in algorithms:
    for size in sizes:
        learn ,deploy = evaluate(alg,size=size)
        results.append([size,'learn',alg.algorithm, learn[-1]/rnd_ctr])
        results.append([size,'deploy',alg.algorithm, deploy[-1]/rnd_ctr])
        
        plt.plot(deploy[100:5000], label="size={}".format(size))
        
    plt.title(alg.algorithm)
    plt.xlabel("T")
    plt.ylabel("CTR")
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()
```

```python
df = pd.DataFrame(results)
df.columns = df.iloc[0]
df = df[1:]
df
```
